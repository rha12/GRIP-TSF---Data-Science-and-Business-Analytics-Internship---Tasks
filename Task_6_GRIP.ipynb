# Prediction using Decision Tree Algorithm
* Create the Decision Tree classifier and visualize it graphically
* Dataset link: https://bit.ly/3kXTdox

## Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.tree import plot_tree,DecisionTreeClassifier
```

## Load the dataset

```python
df = pd.read_csv('Iris.csv')
```

```python
# display first 5 examples
df.head(5)
```

## Data preprocessing

```python
# check for null values
df.isnull().sum()
```

```python
# drop the column Id
df.drop('Id',axis=1,inplace=True)
```

```python
df
```

```python
# split dataset into X,y
features = ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
X,y= df[features], df['Species']
```

## Build Model

```python
# create instance of the descision tree classifier
model= DecisionTreeClassifier(random_state=0)
```

```python
# train the model
model.fit(X,y)
```

## Visualization the Decision Tree

```python
# visualize the descision tree
target=['setosa', 'versicolor', 'virginica']

fig = plt.figure(figsize=(25,20))
_ = plot_tree(model,
                   feature_names= features,
                   class_names= target,
                   filled=True)
```

## Conclusion
### We can now feed any new data to this model ,and it would be able to make accurate predictions.
